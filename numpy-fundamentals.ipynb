Array Attribute

```python
a1 = np.arange(10)
a2 = np.arange(12,dtype=float).reshape(3,4)
a3 = np.arange(1,9).reshape(2,2,2)

print(a1)
print(a2)
print(a3)
```

```python
# ndim --> no of dimensions
a3.ndim
```

```python
# shape
a3.shape
print(a3)
```

```python
# size --> no of elements
a3.size
```

```python
# itemsize --> size in memory
a1.itemsize
```

```python
# dtype
print(a1.dtype)
print(a2.dtype)
print(a3.dtype)
```

```python
# astype --> changing datatype
a3.dtype
a3.astype(np.int32)
# previously it was 64bits and now it is 32bits
```

Array Operations

```python
a1 = np.arange(12).reshape(3,4)
a2 = np.arange(12,24).reshape(3,4)

print(a1)
print(a2)
```

```python
## scalar operations

# arithmetic
a1+2
```

```python
# relational
a2 > 15
```

```python
## vector operations
# arithmetic
a1 + a2
```

Array Functions

```python
a1 = np.random.random((3,3))
a1 = np.round(a1*100)

print(a1)
```

```python
# max/min/sum/prod
# 0 -> col and 1 -> row
np.prod(a1)
np.max(a1,axis=1) # to find max in each row
```

```python
# mean/median/std/var
np.var(a1,axis=1)
```

```python
# trigonometric functions
np.sin(a1)
```

```python
# dot product --> 2 matrix (3,4) and (4,3)
a2 = np.arange(12).reshape(3,4)
a3 = np.arange(12,24).reshape(4,3)

np.dot(a2,a3)
```

```python
# log and exponent
np.exp(a1)
```

```python
# round/floor/cell
np.round(np.random.random((2,3))*100)
```

*Indexing and Slicing*



```python
a1 = np.arange(10)
a2 = np.arange(12).reshape(3,4)
a3 = np.arange(8).reshape(2,2,2)

a3
```

```python
print(a2)
# to locate 6 in a2 array
a2[1,2] # [row,colomn]
```

```python
print(a3)
# to locate 5 in a3 (3d array--tenser)
a3[1,0,1] # [matrix,row,colomn]
```

```python
## slicing
```

```python
a1
```

```python
a1[2:5] # 2 to 5
```

```python
a2
```

```python
# for 1st row in 2d
a2[0,:] # ':' means all columns
```

```python
# 3rd column
a2[:,2] # ':' means all rows
```

```python
# for 5,6,9,10
a2[1:,1:3]
```

```python
# for 0,3,8,11
a2[::2,::3]
```

```python
# for 1,3,9,11
a2[::2,1::2]
```

```python
# for 4 and 7
a2[1,0::3]
```

```python
# for 1,2,3,5,6,7
a2[0:2,1:]
```

```python
## 3d array
a3 = np.arange(27).reshape(3,3,3)

a3
```

```python
print(a3[0,1]) # for 1 matrix middle row
print(a3[1,:,1]) # for 2 matrix middle column
print(a3[2,1:,1:]) # for 22,23,25,26
```

```python
# for 0,2,18,20
a3[::2,0,::2]
```

*Stacking*

```python
# horizontal stacking
a4 = np.arange(12).reshape(3,4)
a5 = np.arange(12,24).reshape(3,4)
```

```python
np.hstack((a4,a5))
```

```python
# vertical stacking
np.vstack((a4,a5))
```
